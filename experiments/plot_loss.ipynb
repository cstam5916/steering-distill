```python
import json
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
with open("../checkpoints/steering-final/checkpoint-8424/trainer_state.json") as f:
    state = json.load(f)

logs = state["log_history"]
NUM_TRAIN = 7473
NUM_EVAL = 1319

train_epochs = [x["epoch"] for x in logs if "loss" in x]
train_loss = [x["loss"] / NUM_TRAIN for x in logs if "loss" in x]

val_epochs = [x["epoch"] for x in logs if "eval_loss" in x]
val_loss = [x["eval_loss"] / NUM_EVAL for x in logs if "eval_loss" in x]

plt.plot(train_epochs, train_loss, label="Train loss")
plt.plot(val_epochs, val_loss, label="Validation loss")

plt.title('Loss curves for SVKD loss')
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend()
plt.show()
```
